### Data query
To detect cascading failures of delays, data set has been filtered to only include non-cancelled and flights that has departure delays more than 0 minutes and calculates its average departure delays. 

```python
q7 = c.execute('''   
                SELECT AVG(DepDelay) as avg_depdelay, Dest, Month, Origin, DayofMonth, Year, DepTime, CRSDepTime
                FROM ontime JOIN airports ON Dest=iata 
                WHERE Cancelled = 0 AND DepDelay > 0
                GROUP BY Dest, Month, Year, DayofMonth, DepTime, CRSDepTime
                ORDER BY avg_depdelay ASC
                  ''').fetchall()
q7 = pd.DataFrame(q7).reset_index()

q7.columns = ['index','avg_depdelay', 'dest', 'month', 'origin', 'dayofmonth','year','deptime','crsdeptime']
q7 = q7.drop(columns=['index'])
```

To analyze in-depth on cascading failures , data has been filtered to compare the average departure delays on 1st day of the month for 2007 and 2008. Percentage of departure delays has been calculated to reflect and highlight the weightage of individual delays over total delays on 1st day of the month for both years incurred from flights between origin to destination.

```python
q7_a = q7[(q7['dayofmonth']==1) & (q7['month'] ==1) & (q7['year']==2007)]
q7_a = q7_a.nlargest(100,'avg_depdelay')

q7_b = q7[(q7['dayofmonth']==1) & (q7['month'] ==1) & (q7['year']==2008)]
q7_b = q7_b.nlargest(100,'avg_depdelay')

q7_a['pct_avg_depdelay'] = q7_a['avg_depdelay']/q7_a.groupby('dest')['avg_depdelay'].transform('sum')
q7_b['pct_avg_depdelay'] = q7_b['avg_depdelay']/q7_b.groupby('dest')['avg_depdelay'].transform('sum')
```

Plot network graph to show cascading failures of departure delays affecting other airports from origin to destination on the 1st day of the month. Isolated Nodes (airports) with no degrees or connection will be removed to visualize cascading failures. Additionally, percentage of average departure delay has been used as weights to determine the impact of each individual delays over total delays on 1st day of the month for 2007 and 2008.

```python
import networkx as nx
G_2007 = nx.Graph()

for index, row in q7_a.iterrows():  
    G_2007.add_edge(row['origin'], row['dest'], weight=row['pct_avg_depdelay']) 
    
remove = [node for node, degree in G_2007.degree() if degree==0]  
G_2007.remove_nodes_from(remove) 

# Visualize network via networkx packages.
%matplotlib inline

options = {
     'node_color': 'black', 
     'edge_color': 'purple',
     'node_size': 1, 
     'alpha': 0.5 

}

plt.subplots(figsize=(10,8))
pos = nx.circular_layout(G_2007) 

edges = G_2007.edges() 
weights = [G_2007[u][v]['weight'] for u,v in edges] # using weights (percentage_avg_depdelay) as u and v edges. 

nx.draw_networkx_labels(G_2007,pos=pos,font_size=9)
nx.draw(G_2007,pos=pos,font_size=9, width = weights, **options)

plt.tight_layout()
plt.axis('off'); # hide axis.
plt.show() 


```

```python
G_2008 = nx.Graph()
for index, row in q7_b.iterrows(): 
    G_2008.add_edge(row['origin'], row['dest'], weight=row['pct_avg_depdelay'])
    
remove = [node for node, degree in G_2008.degree() if degree==0] 
G_2008.remove_nodes_from(remove)

%matplotlib inline

options = {
     'node_color': 'black', 
     'edge_color': 'purple',
     'node_size': 1,
     'alpha': 0.5 

}

plt.subplots(figsize=(10,8))
pos = nx.circular_layout(G_2008) 
edges = G_2008.edges() 
weights = [G_2008[u][v]['weight'] for u,v in edges] # using weights(percentage_avg_depdelay) as u and v edges. 
nx.draw_networkx_labels(G_2008,pos=pos,font_size=9)
nx.draw(G_2008,pos=pos,font_size=9, width = weights, **options)

plt.tight_layout()
plt.axis('off'); 
plt.show()
```

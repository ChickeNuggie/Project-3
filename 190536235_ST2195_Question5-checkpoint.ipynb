### Load the library

```python
import sqlite3
import os
import seaborn as sns 
import random
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import train_test_split, GridSearchCV      
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer 
from xgboost.sklearn import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, accuracy_score
from math import sqrt
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Ridge
from sklearn import linear_model
import seaborn as sb
```

### Data Query
To determine variables that are useful in predicting departure delay model, correlation matrix has been drawn to visualize individual variables that are correlated with departure delays. Due to computer constraints, data set has been further sampled into n=20,000 from previous 30% sampled data for efficient performance. In order to run machine learning algorithms, cancelled and diverted flights which contains NA values will be removed from the data to ensure smooth performance.     ***Note: Results may different due to random subset generated***

```python
q8 = c.execute('''   
               SELECT *
               FROM ontime 
               WHERE Cancelled = 0 AND Diverted =0 
               ''').fetchall()
c_name_list = [tuple[0] for tuple in c.description]

q8 = pd.DataFrame(q8).reset_index()
q8 = q8.drop(columns=['index'])
q8.columns = [c_name_list]
q8.isnull().sum()
q8.dtypes
q9 = q8.sample(n=20000)

q9_corr = q9.corr()
#sns.heatmap(q9_corr, 
#        xticklabels=q9_corr.columns,
#        yticklabels=q9_corr.columns, annot_kws={"size": 7},fmt=".1f",linewidth=1)
sb.heatmap(q9_corr, cmap="Blues")
```

Below are the final selected predicators to predict depdelay.

```python
features = ['CRSDepTime','CRSArrTime','CarrierDelay','WeatherDelay','NASDelay','LateAircraftDelay']
X = q9[features].copy()
y = q9['DepDelay'].copy()
X.columns = ['CRS_Dep_Time','CRS_Arr_Time','Carrier_Delay','Weather_Delay','NAS_Delay','Late_Aircraft_Delay']
```

To ensure prediction model is generated, data has been imputed using mean and median for numeric or integers data types to facilitate comparison and find the highest accuracy via lowest Root Mean Square Error (RMSE), Mean Square Error (MSE), Mean Absolute Error (MAE) and highest R^2 for regression among machine learning learning algorithms. Additionally, scaling is used to ensure equal measurement unit, resulting in more accurate prediction. These will and combined into a graph learner before model evaluation.

```python
numerical_features = ['CRS_Dep_Time','CRS_Arr_Time','Carrier_Delay','Weather_Delay','NAS_Delay','Late_Aircraft_Delay']
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer()),
    ('scaler', StandardScaler())])

data_transformer = ColumnTransformer(
    transformers=[
        ('numerical', numerical_transformer, numerical_features)])

param_grid = {
    'data_transformer__numerical__imputer__strategy': ['mean', 'median'] 
}
```

70% Train and 30% test set split is used to determine how well the model predicts future outcome with test set(unseen model) based on training observations (seen model). 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=1)
```

#### Pipeline with Linear Regression model into GraphLearner
10-fold cross validation has been used to ensure the models are being tested thoroughly to improve overall prediction accuracy.

```python
lm = Pipeline(steps=[('data_transformer', data_transformer),
                          ('lm', LinearRegression())])

grid_lm = GridSearchCV(lm, param_grid=param_grid, cv=10)
grid_lm.fit(X_train, y_train);  

print(grid_lm.best_score_)  #mean cross-validated score of the best_estimator
print(grid_lm.best_params_)
print(grid_lm.best_estimator_)
grid_lm.cv_results_

y_predict_lm = grid_lm.predict(X_test)

RMSE_lm= np.sqrt(mean_squared_error(y_test, y_predict_lm))
MSE_lm= mean_squared_error(y_test, y_predict_lm)
MAE_lm= mean_absolute_error(y_test, y_predict_lm)
r2_lm= r2_score(y_test, y_predict_lm)

print('RMSE =', RMSE_lm) # RMSE = 10.52
print('MSE =', MSE_lm) # MSE = 110.65
print('MAE =', MAE_lm) # MAE = 6.50
print('R2 =', r2_lm) # R2 = 0.87
```

Plot the graph to show predicted against actual departure delay generated from linear regression model.

```python
sns.regplot(x=y_predict_lm, y=y_test, scatter_kws={"color": "purple"}, line_kws={"color": "red"}, marker="^")
plt.xlabel("Predicted Departure Delay")
plt.ylabel("Actual Departure Delay")
plt.title("Linear Regression Model")
plt.show()
```

#### Pipeline with Ridge Regression model into GraphLearner

```python
lmr = Pipeline(steps=[('data_transformer', data_transformer),
                          ('lmr', Ridge(alpha=50))]) 

grid_lmr = GridSearchCV(lmr, param_grid=param_grid, cv=10)
grid_lmr.fit(X_train, y_train.values.ravel());

print(grid_lmr.best_score_) 
print(grid_lmr.best_params_)
print(grid_lmr.best_estimator_) 
 
y_predict_lmr = grid_lmr.predict(X_test)

RMSE_lmr= np.sqrt(mean_squared_error(y_test, y_predict_lmr))
MSE_lmr= mean_squared_error(y_test, y_predict_lmr)
MAE_lmr= mean_absolute_error(y_test, y_predict_lmr)
r2_lmr= r2_score(y_test, y_predict_lmr)

print('RMSE =', RMSE_lmr) # RMSE= 10.51
print('MSE =', MSE_lmr) # MSE= 110.46
print('MAE =', MAE_lmr) # MAE = 6.51
print('R2 =', r2_lmr)  # R^2 = 0.87
```

Plot the graph to show predicted against actual departure delay generated from Ridge regression model.

```python
sns.regplot(x=y_predict_lmr, y=y_test, scatter_kws={"color": "green"}, line_kws={"color": "red"}, marker = "D")
plt.xlabel("Predicted Departure Delay")
plt.ylabel("Actual Departure Delay")
plt.title("Ridge Linear Regression")
plt.show()
```

#### Pipeline with Lasso Regression model into GraphLearner

```python
lml = Pipeline(steps=[('data_transformer', data_transformer),
                      ('lml', linear_model.Lasso(alpha=0.1))]) 

grid_lml = GridSearchCV(lml, param_grid=param_grid, cv=10)
grid_lml.fit(X_train, y_train);

print(grid_lml.best_score_) 
print(grid_lml.best_params_)
print(grid_lml.best_estimator_) 
 
y_predict_lml = grid_lml.predict(X_test)

RMSE_lml= np.sqrt(mean_squared_error(y_test, y_predict_lml))
MSE_lml= mean_squared_error(y_test, y_predict_lml)
MAE_lml= mean_absolute_error(y_test, y_predict_lml)
r2_lml= r2_score(y_test, y_predict_lml)

print('RMSE =', RMSE_lml) # RMSE= 10.51
print('MSE =', MSE_lml) # MSE= 110.45
print('MAE =', MAE_lml) # MAE = 6.5
print('R2 =', r2_lml)  # R^2 = 0.87
```
